```python
# how to merge tow diff. table 
```

```python
table1=pd.DataFrame({'name':['dipesh','raj','viaml','rahul'],
        'erp':[12,34,56,78],
        'roll no':[45,123,67,879],
        'pclass':['a','b','c','d']
})
```

```python
table2=pd.DataFrame({'name':['rakesh','ravan','nimesh','sagar','sdhu'],
        'erp':[123,345,567,89,67],
        'roll no':[567,98,421,568,69],
        'pclass':['e','f','g','h','k']
})
```

```python
pd.concat([table1,table2])           # two table merge in same direction 
```

```python
pd.concat([table1,table2],axis=1)   # two table merge side by side
```

```python
data1=pd.DataFrame({'key1': [1,2,4,5,6],
                    'key2': [4,5,6,7,8],
                    'key3': [3,4,5,6,6]})
```

```python
data2=pd.DataFrame({'key1':[1,2,45,6,67],
                    'key4':[4,5,6,7,8],
                    'key5':[3,4,5,6,6]})
data1
```

```python
data2
```

```python
pd.merge(data1,data2)  # by default is inner
```

```python
pd.merge(data1,data2 ,how='left')
```

```python
pd.merge(data1,data2 ,how='right')
```

```python
pd.merge(data1,data2,how='outer')
```

```python
pd.merge(data1,data2,how='cross')  # creates the cartesian product from both frames, preserves the order of the left keys.
```

```python
data3=pd.DataFrame({'key1': [1,2,4,5,6],
                    'key2': [4,5,6,7,8],
                    'key3': [3,4,5,6,6]},
                   index=['a','b','c','d','e']
                  )
data4=pd.DataFrame({'key4':[1,2,45,6,67],
                    'key5':[4,5,6,7,8],
                    'key6':[3,4,5,6,6]},
                   index=['f','i','j','k','l']
                  )
```

```python
# Both join and merge can be used to combines two dataframes but the join method combines two dataframes on the basis of their indexes
# the merge method is more versatile and allows us to specify columns to join on for both dataframes.
```

```python
data3.join(data4)   #  by default is left 
```

```python
data3.join(data4,how='outer')
```
